```python
import requests
import pandas as pd  # Import pandas library

def fetch_quotes():
    url = "https://apidojo-yahoo-finance-v1.p.rapidapi.com/auto-complete"
    querystring = {"q": "tesla", "region": "US"}

    headers = {
        "X-RapidAPI-Key": "aa48581c80msh6dc13c85a681dc4p143375jsn17249c4b6104",
        "X-RapidAPI-Host": "apidojo-yahoo-finance-v1.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)
    quotes = response.json().get('quotes', []) # Extracting data from response
    df = pd.DataFrame(quotes)
    return df

dfs = [fetch_quotes() for i in range(50)]    # iterating 50 times throught function and appends all dataframes
final_df = pd.concat(dfs, ignore_index=True)

final_df  #result
```
